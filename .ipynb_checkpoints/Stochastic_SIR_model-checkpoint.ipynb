# SIR model via Gillespie algorithm

```python
import pandas as pd, holoviews as hv, networkx as nx, random as rm, panel as pn
from holoviews import opts
from decimal import *
import hvplot.networkx as hvnx
import hvplot.pandas
hv.renderer('bokeh').theme = 'light_minimal' # Graph style
pn.extension()
pd.options.plotting.backend='holoviews'
```

## The SIR model

The SIR model is one of the simplest models for simulating the spread of a disease. The population is divided into three classes: 
- Susceptible (S): people who can get infected
- Infected (I): people who are now ill
- Recovered (R): people who have recovered from the disease

It is only possible to move from S to I and from I to S. Here we will assume that a person who has recovered is immune and cannot get the disease anymore.

The equations that describe the rate of change of the number of susceptible, infected and recovered can be written as:

$$
\begin{align}
\frac{dS}{dt} &= -\beta\frac{I S}{N}\\
\frac{dI}{dt} &= \beta\frac{I S}{N} - \gamma I\\
\frac{dR}{dt} &= \gamma I
\end{align}
$$

Where $\beta$ is the infection rate, $\gamma$ is the recovery rate and $N$ is the total number of people.

A useful quantity for evaluating the impact of a disease is the *basic reproduction number*, which is defined as:
$$R_0 = \frac{\beta}{\gamma}$$
The equation for the infection rate can be written in terms of this quantity as:
$$
\frac{dI}{dt} = \left(R_0\frac{ S}{N}-1\right) \gamma I
$$
Therefore, the following cases are possible:
$$
\begin{align}
R_0&>\frac{N}{S(t=0)}\approx 1 \rightarrow \frac{dI}{dt}>0 \rightarrow \textrm{exponential outbreak}\\
R_0&<\frac{N}{S(t=0)}\approx 1 \rightarrow \frac{dI}{dt}<0 \rightarrow \textrm{exponential decay}
\end{align}
$$

In the first case the disease is called *epidemic*.

## The Gillespie algorithm applied to the SIR model

A simulation of a SIR process can be achieved using the Gillespie algorithm. When applied to this model, the Gillespie algorithm runs as follows:

1. For each person, find the rate of change of their state, then calculate the **total rate of change** as the sum of the rates of all people. If a person is infected, their rate of change is $\gamma$, if instead they are susceptible, their rate of change is $m\cdot\beta$, where $m$ is the number of infected people they are in contact with.
2. Find the **time** when the **next event** (infection or recovery) will occur. This is done by extracting the time from an exponential distribution with rate equal to the total rate of change.
3. Select **which event occurs**, according to the probability of each event's rate. This process can be divided into two parts <br>
    - *Determine if the event is a recovery or an infection*: The total rate $R$ can be written as the sum of the total infection rate $R_{inf}$ and total recovery rate $R_{rec}$. Therefore, for this step we extract a random number between 0 and $R$, if it's lower than $R_{rec}$, a recovery occurs, if it's higher an infection occurs.<br>
    - *Determine which node gets recovered/infected*: If the event is a recovery, since the recovery rate is $\gamma$ for all infected people, we randomly choose one of them. Instead, if event is an infection, we need to consider that some suscecptible nodes have an infection rate higher than others because they have more infected neighbors. For this reason, we track the maximum infection rate $r_{max}$ and take the ratio of the rate of a randomly extracted person $r_n$ and $r_{max}$. We extract another random number $p$ between 0 and 1 and if $p<r_n/r_{max}$, $n$ gets infected, otherwise we extract another person $n'$ and repeat the process.
4. **Update the rates** according to which person has changed their state and repeat the process.

The algorithm stops when $R = 0$

<br>

The following function runs this algorithm, given a network describing the interactions between the memembers of the population, the recovery rate and $R_0$.

```python
def SIR_Gillespie(K, gamma, R0, graphs = False, info = False):
    '''
    K: Network
    gamma: recovery rate
    R0: ratio between infection and recovery rate
    graphs: True if you want a dictionary of graph plots, one for each moment in time
    info: True to print debugging info while the function runs
    '''
    
    beta = gamma*R0             # infection rate
    N = K.number_of_nodes()     # Number of nodes in network

# Set initial conditions (State = S or I and Rate = beta or gamma)
    K.nodes[0]['State'] = 'I'                      # Set node 0 as infected
    K.nodes[0]['Rate'] = gamma                     # Set rate of node 0 to the recovery rate
    for i, node in K.nodes.data():                 # Set all other nodes as 'S'
        if i == 0: continue                        # Skip first node, which is 'I'
        node['State'] = 'S'                         
        if i in K[0]: node['Rate']=beta    # Neighbours of infected node have rate = beta
        else: node['Rate'] = 0             # Other nodes have infection rate 0
    
# Lists of infected/susceptible/recovered/at risk nodes
    I = [0]                         
    S = [i for i in range(1, N)]
    R = []
    At_risk = list(K[0])   # At the beginning the at risk nodes are the neighbours of the infected node
    
# Dataframe containing S,I and R at each point in time (for final plot)
    SIR = pd.DataFrame(index = [0], data = {'Susceptible':len(S), 'Infected':len(I), 'Recovered':len(R)})
    
# Graphs setup, for displaying the network evolution over time
    if graphs == True:
        S_col, I_col, R_col = '#30a2da', 'orangered', 'mediumseagreen'     # Colors of S, I and R nodes
        cmap = [I_col] + [S_col]*(N-1)                                     # Colormap that is updated as S, I and R change
        pos = nx.spring_layout(K)                                          # Define the position of all nodes, so that all graphs have the same layout
        opts_graph = {'node_color': cmap, 'pos': pos,                      # Graph esthetics
                      'node_size' : 70, 'edge_width' : .5, 
                      'width' : 400, 'height' : 400}
        Graphs = {0: hvnx.draw(K, **opts_graph)}                           # List containing the graph plot for each time
    
# Setting of initial rates and times
    max_rate = beta                                    # Maximum inf. among all at risk nodes (r* in book)
    tot_rec_rate = gamma*len(I)
    tot_inf_rate = beta*len(At_risk)  
    tot_rate = tot_rec_rate + tot_inf_rate    
    time = rm.expovariate(float(tot_rate))             # Time of the first event, chosen from exp distribution
        
    j = 0                                              # Counter for the  number of loops
 
 # Main loop

    while True:
                
        r1 = rm.uniform(0, float(tot_rate))                          # Extract random number
            
    # Recovery
        if r1<tot_rec_rate and j!=0 and len(I)>0:                     # In the first loop we force the infection, to avoid the epidemic to stop immediately.
            r2 = rm.choice(I)                                         # Choose a random infected node
            I.remove(r2)                                              # Remove node from infected
            R.append(r2)                                              # Add node to recovered
            K.nodes[r2]['State'] = 'R'                                # Change its state to recovered
            K.nodes[r2]['Rate'] = 0                                   # Change its rate to 0
            suscep_neighbors = [n for n in K.neighbors(r2)            # Neighbors of node r2 that are susceptible    
                                if K.nodes[n]['State'] == 'S']
            for nbr in suscep_neighbors:                              # Decrease the rate of the susc. neighbors since this node recovered
                K.nodes[nbr]['Rate'] -= beta
                if K.nodes[nbr]['Rate'] == 0:                         # If the neighbor has no other I neighbors (i.e. rate = 0) 
                    At_risk.remove(nbr)                               # Remove it from at risk list
            tot_rec_rate -= gamma                                     # The total rec. rate decreases by 1 since I decreases by 1
            tot_inf_rate -= len(suscep_neighbors) * beta              # The total infection rate decreases by beta for every susc. neighbor
            if graphs == True: cmap[r2] = R_col
            
    #Infection
        else:
            if len(At_risk)>0:                                        # We do all of this only if there are still infected people.
                max_rate = max(K.nodes[n]['Rate'] for n in At_risk)
                while True:                                           # Choose the node to be infected
                    r2 = rm.choice(At_risk)                           # Choose a random at risk node
                    p = rm.random()                                   # Choose a random number in range (0,1)
                    if p<(K.nodes[r2]['Rate']/max_rate): break        # If this condition is satisfied, this is the node that gets infected
                suscep_neighbors = [n for n in K.neighbors(r2)        # Neighbors of node r2 that are susceptible    
                                    if K.nodes[n]['State'] == 'S']
                for nbr in suscep_neighbors:
                    K.nodes[nbr]['Rate'] += beta                      # Increase the rate of all S neighbor by beta
                    if nbr not in At_risk:                            # Make sure all S neighbors are at risk
                        At_risk.append(nbr)
                tot_rec_rate += gamma                                 # The total rec. rate increases by 1 since I increases by 1
                tot_inf_rate += len(suscep_neighbors) * beta          # The total infection rate increases by beta for every susc. neighbor        
                tot_inf_rate -= K.nodes[r2]['Rate']                   # Remove the rate of the infected node, since now it only contributes to recovery rate
                S.remove(r2)                                          # Remove node from susceptibles
                I.append(r2)                                          # Add node to infected
                K.nodes[r2]['State'] = 'I'
                K.nodes[r2]['Rate'] = float(gamma)
                At_risk.remove(r2)                                    # Node is not at risk anymore
                if graphs == True: cmap[r2] = I_col
            
    # Updates
        tot_rate = tot_rec_rate + tot_inf_rate
        if info and tot_inf_rate<0: print('NEGATIVE INF RATE!')
        if info and tot_rec_rate<0: print('NEGATIVE REC RATE!')
        SIR.loc[time] = [len(S), len(I), len(R)]                      # Update SIR DataFrame
        if graphs == True: 
            Graphs[time] = hvnx.draw(K, **opts_graph)                 # Add the updated graph to the dictionary of graphs
        if tot_rate>0: time += rm.expovariate(float(tot_rate))        # I go on by a time step with exp distribution
        else: break                                                   # If the rate is null, I'm done
        j += 1
        
    if graphs == False: return SIR
    else: return SIR, Graphs
```

## Simulations

We consider a population of 100 people, arranged into a network where each node represents a person and each edge between nodes represents a relationship between two people. We use a **stochastic block model** network, where the population is divided into different groups. Within each group the probability of having a relationship between two nodes is $P_{in}$, while the probability for two nodes belonging to different groups is $P_{out}$. 
The following table summarizes the parameters chosen for this simulation:

|N|Groups|$P_{in}$|$P_{out}$|
|-------:|:-------:|:-------:|:-------:
|100|3|40%|1%|

This configuration is used to simulate different communities inside the population, with most contacts happening inside the groups and some rarer contacts between individuals of different groups.

```python
N = 100                                        # Total number of people
Groups = 3                                     # Number of groups of people
prob_in = 0.4                                  # probability of an edge between nodes of the same group
prob_out = 0.01                                # probability of an edge between nodes of different groups
sizes = [N//Groups for i in range(Groups)]     # Split all nodes equally between the groups
probs = [[prob_in] + [prob_out]*(Groups-1)]    # Create symmetric matrix for in and out probabilities
for i in range(1, Groups):
    probs.append(probs[0][-i:]+probs[0][:-i])
```

### The effect of R0 with graphs visualization

Here we run the model with two different values of $R_0$, keeping the value of the recovery rate fixed at $\gamma = 1$. In the first case, we set $R_0 = 0.2$, for which the spred is expected to die out quickly. In the second case we set an epidemic reproduction number $R_0 = 2$.

The plots below show the time evolution of the number of susceptible, infected and recovered. To each plot is associated an interactive graph that allows to see how the spread evolved over time by moving a slider. The blue dots are S nodes, red ones are I nodes and green ones are R nodes

```python
K_1 = nx.stochastic_block_model(sizes, probs)    # Create network 1
K_2 = nx.stochastic_block_model(sizes, probs)    # Create network 2

gamma = 1
R0_1 = 0.1
R0_2 = 2

SIR_1, Graph_1 = SIR_Gillespie(K=K_1, gamma = gamma, R0 = R0_1, graphs=True)
SIR_2, Graph_2 = SIR_Gillespie(K=K_2, gamma = gamma, R0 = R0_2, graphs=True)
```

```python
def Plot_SIR(SIR, title = ''):
    cmap = hv.Cycle(['orangered', '#30a2da', 'mediumseagreen'])
    plt_SIR = hv.Area.stack(SIR['Infected'].plot()*SIR['Susceptible'].plot()*SIR['Recovered'].plot())
    plt_SIR.opts(width = 700, height = 400, padding = 0, legend_position = 'right', 
             xlabel = 'Time', ylabel = 'Counts', shared_axes = False, title = title)
    plt_SIR.opts(opts.Area(color = cmap))
    return plt_SIR

pn.Column(pn.Row(pn.Spacer(width = 250), '## R0 = 0.2'), Plot_SIR(SIR_1, title = 'Figure 1'), 
          hv.HoloMap(Graph_1, kdims = 'Time'),
          pn.Row(pn.Spacer(width = 250), '## R0 = 2'), Plot_SIR(SIR_2, title = 'Figure 2'), 
          hv.HoloMap(Graph_2, kdims = 'Time'))
```

#### Discussion
These simulations show the big effect that different values of $R_0$ can have on the spread of a disease. 
As expected from the theory, if $R_0>1$ the infection rapidly spreads infecting the whole population (Figure 2). Instead, if $R_0<1$ the disease quickly fades out (Figure 1). 

### Dependence of the number of infected on $R_{0}$ 

The main way a population can minimize $R_0$ is by trying to decrease $\beta$, for example with social distancing and better personal hygiene. Generally $\gamma$ is fixed, since it is related to the virus itself and the average time needed for an organism to recover.

Here we see how the number of infected changes in relation to $R_0$, in order to understand what is the effect social distancing and other precautions can have on the national health system.

We fix the recovery rate at $\gamma = 1$ and track how changing $R_0$ affects the following quantities:
- Total number of infected people
- Maximum number of infected in a given day

The second quantity is also relavant because, if this value is too high, the national health system might not be able to take care of every infected person due to lack of doctors or places inside intensive care.

We chose the following values for $R_0$:
$$R_0 = \{0.2,0.5,1,2\}$$
For each $R_0$, we run the simulation *800 times* and then display on a histogram the counts for the two quantities written above.

```python
K = nx.stochastic_block_model(sizes, probs)    # Create network
getcontext().prec = 5                          # Set decimal precision to 5

N_rep = 800
gamma = 1
R0_lst = [2, 1, 0.5, 0.2]
columns = ['R0 = '+str(i) for i in R0_lst]

Itot_R0, Imax_R0 = pd.DataFrame(columns = columns), pd.DataFrame(columns = columns)
for i in range(N_rep):   
    tots, maxs = [], []
    #print(i+1, '/', N_rep, end = '\r')
    for j, R0 in enumerate(R0_lst):
        SIR = SIR_Gillespie(K, gamma = Decimal(str(gamma)), R0 = Decimal(str(R0)))
        tots.append(N-SIR.Susceptible.iloc[-1])
        maxs.append(SIR.Infected.max())
        if i == 0 and R0 == 2: 
            I_2 = SIR.Infected   
        if i == 0 and R0 == 0.5: 
            I_0p5 = SIR.Infected                    
    Itot_R0.loc[i] = tots        # Tot infected = # of nodes - Susceptible at the end
    Imax_R0.loc[i] = maxs
```

```python
# Plot
options = {'bins': 90, 'shared_axes': False, 'height' : 400, 'ylabel': 'Counts', 'alpha': 0.8}
(Itot_R0.plot.hist(xlabel = 'Total infected',
                 label = 'Figure 3: Total number of infected people', **options) + \
 Imax_R0.plot.hist(xlabel = 'Maximum of daily infected', 
                 label = 'Figure 4: Maximum number of infected per day',**options)).cols(1)
```

#### Discussion

These results help understand how a change in $R_0$ can help the National Health System during the spread of a disease.
It is interesting that even if $R_0 = 0.5$, most of the times all the population gets infected. With $R_0 = 0.2$ there is instead a big spread in the results, ranging from 0% to more than 90% of the population (Figure 4), but also in this case there is an almost 50% chance that more than 80% of the population gets infected.  <br>
However, what changes drastically with $R_0$ is the maximum number of infected people in a given day (Figure 4). If $R_0 = 2$, in the day where the number of infected is maximum, between 70 and 90% of the population is expected to be infected. This can be a big problem for the National Health System, which might not be able to keep up with such a high healthcare request. Instead,as $R_0$ decreases this percentage gets lower, for example in the case of $R_0 = 0.2$, this drops between 1 and 40%. 

Therefore, with a lower $R_0$, there is still a high chance that the whole population gets infected, but in a longer time, so that the Health System is able to cure everyone during the epidemic. This is demonstrated in Figure 5, which represents the curves of the number of infected people as a function of time for $R_0 = 2$ and $R_0 = 0.5$. The curve with lower $R_0$ is more spread out and has a lower maximum value

```python
(I_2.plot(label = 'R0 = 2')*\
 I_0p5.plot(label = 'R0 = 0.5')).opts(width = 600, height = 400, 
                                      xlabel = 'Time', ylabel = 'Number of infected', 
                                      title = 'Figure 5')
```
